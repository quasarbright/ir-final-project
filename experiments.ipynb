```python
from bs4 import BeautifulSoup, Tag
import requests
from urllib.parse import urljoin
import re
```

# crawl

```python
toc_url = "https://docs.racket-lang.org/reference/index.html"
resp = requests.get(toc_url)
reference_html = resp.text
reference_html
```

```python
soup = BeautifulSoup(reference_html, 'html.parser')
```

```python
main = soup.body.find_next("div", {"class": "main"})
table = main(string="Language Model")[0].parent.parent.parent.parent.parent
toc_links = table.find_all('a')[:-2]

urls = [urljoin(toc_url, link['href']) for link in toc_links]
urls
```

```python
hashless_urls = [url for url in urls if "#" not in url]
hashless_urls
```

```python
len(hashless_urls)
```

# scrape

```python
eval_url = "https://docs.racket-lang.org/reference/eval-model.html"
resp = requests.get(eval_url)
eval_html = resp.text
eval_soup = BeautifulSoup(eval_html, 'html.parser')
eval_soup
```

```python
main = eval_soup.find('div', class_='main')
children = list(main.children)
children = [child for child in children if child.name != 'div']
[child.name for child in children]
isinstance(eval_soup, Tag)
```
